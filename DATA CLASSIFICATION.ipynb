```python
emp.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis=1,inplace=True)
```

```python
emp.columns
```

```python
emp.loc[emp.Attrition=='Yes','Attrition']=1
emp.loc[emp.Attrition=='No','Attrition']=0
```

```python
emp.Attrition.unique()
```

```python
emp=pd.get_dummies(emp,drop_first=True)
```

```python
emp.shape
```

```python
emp.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
emp_model=emp.copy()
```

```python
emp_model['Attrition_1'].head()
```

```python
emp.rename(columns={'Attrition_1':'Attrition'},inplace=True)
```

```python
emp_model['Attrition_1'].value_counts()
```

```python
train_X=emp_model.drop(['Attrition_1'],axis=1)
train_Y=emp_model['Attrition_1']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(train_X,train_Y,test_size=0.2)
```

```python
print(X_train.shape)
```

```python
from sklearn import tree
```

```python
dt=tree.DecisionTreeClassifier()
```

```python
dt.fit(X_train,Y_train)
```

```python
Y_pred=dt.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print("Accuracy - ",metrics.accuracy_score(Y_test,Y_pred))
```
